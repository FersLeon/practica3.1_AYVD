# **Manejo de errores faltantes**

```python
data['Date'] = pd.to_datetime(data['Date'])
df_diario = data.groupby(data['Date'].dt.date).sum().reset_index()
```

```python
#Grafica que resalta los valores nulos
for col in data.columns[1:]:

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(df_diario['Date'], data[col], marker='o')

    # Resaltar los valores nulos con líneas que los encierren
    flag_nulos = data[col].isnull()
    for i in range(len(data)):
        if flag_nulos[i]:
            ax.axvline(x=df_diario['Date'][i], color='red', linestyle='--', linewidth=1)

    ax.xaxis_date()
    plt.xticks(rotation=45)

    ax.set_xlabel('Fecha')
    ax.set_ylabel('Valor')
    ax.set_title('Gráfica de Líneas con Resaltado de Valores Nulos')
    ax.legend()

    plt.show()
```

```python
data_n = data.copy(deep = True)
data_n.isnull().sum()
```

```python
#Llenado con media desplegable
#tiene un periodo de una semana
for col in data_n.columns:
  if data_n[col].isnull().any():
    media_desplegable = data_n[col].rolling(window=30, min_periods=1).mean()
    data_n[col] = data_n[col].fillna(media_desplegable)
```

```python
data.isnull().sum()
```

```python
for col in data_n.columns[1:]:
    fig, ax = plt.subplots(figsize=(10, 6))

    ax.plot(df_diario['Date'], data_n[col], marker='o', label='Datos Cambiados',color='orange')

    ax.plot(df_diario['Date'], data[col], marker='o', label='Datos Originales')

    ax.xaxis_date()
    plt.xticks(rotation=45)

    ax.set_xlabel('Fecha')
    ax.set_ylabel('Valor')
    ax.set_title(f'Comparación de Datos para {col}')
    ax.legend()

    plt.show()
```

# **Suavizado de los datos**

```python
# Obtener el número de instancias (filas)
data.shape[0]
```

```python
import pandas as pd

data['Date'] = pd.to_datetime(data['Date'])
#Indice del dataframe
data = data.set_index('Date')

# Upsample a una frecuencia más alta
df_upsampled = data.resample('H').mean()

df_upsampled = df_upsampled.fillna(method='ffill')  #Relleno hacia adelante (forward fill)

# restablecer Date como una columna en lugar de un índice
df_upsampled = df_upsampled.reset_index()
```
