# **Distancia en el tiempo equidistante**

```python
#Antes de ordenar
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

difechas = data['Date'].diff().dropna()
equidistancia = difechas.dt.total_seconds().value_counts().index.tolist()
if len(equidistancia) == 1:
    print(f"Los datos están equidistantes en el tiempo: {equidistancia[0]} segundos.")
else:
    print("Los datos no están equidistantes en el tiempo.")
```

```python
# Ordenar
datos = data.sort_values('Date')
```

```python
#Despues de ordenar
fechasordenadas = datos['Date'].diff().dropna()
equidistanciaord = fechasordenadas.dt.total_seconds().value_counts().index.tolist()
if len(equidistanciaord) == 1:
    print(f"Después de ordenar, los datos siguen siendo equidistantes en el tiempo: {equidistanciaord[0]} segundos.")
else:
    print("Después de ordenar, los datos ya no son equidistantes en el tiempo.")
```

# **Manejo de valores faltantes y gráficas**

```python
data = pd.read_csv('/content/drive/MyDrive/Aquifer_Petrignano.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
df_diario = data.groupby(data['Date'].dt.date).sum().reset_index()
```

```python
#Grafica que resalta los valores nulos
for col in data.columns[1:]:

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(df_diario['Date'], data[col], marker='o')

    # Resaltar los valores nulos con líneas que los encierren
    flag_nulos = data[col].isnull()
    for i in range(len(data)):
        if flag_nulos[i]:
            ax.axvline(x=df_diario['Date'][i], color='red', linestyle='--', linewidth=1)

    ax.xaxis_date()
    plt.xticks(rotation=45)

    ax.set_xlabel('Fecha')
    ax.set_ylabel('Valor')
    ax.set_title('Gráfica de Líneas con Resaltado de Valores Nulos')
    ax.legend()

    plt.show()
```

```python
import pandas as pd

data_n['Date'] = pd.to_datetime(data_n['Date'])
#Indice del dataframe
data_n = data_n.set_index('Date')

# Upsample a una frecuencia más alta
df_upsampled = data_n.resample('H').mean()

df_upsampled = df_upsampled.fillna(method='ffill')  # relleno hacia adelante (forward fill)

#Restablecer 'Date' como una columna en lugar de un índice
df_upsampled = df_upsampled.reset_index()
```

# **Estacionariedad**

```python
import pandas as pd
import matplotlib.pyplot as plt

columnas_numericas = df_upsampled.select_dtypes(include='number').columns

def prueba_dickey_fuller(serie, nombre_columna):

    media_movil = serie.rolling(window=12).mean()

    # Cálculo de la diferencia entre la serie temporal y su media móvil
    diferencia = serie - media_movil
    diferencia.dropna(inplace=True)

    # Calculando la varianza de la diferencia
    varianza_diferencia = diferencia.var()

    # Calculando la varianza de la serie temporal
    varianza_serie = serie.var()

    # Estimación de la estadística de prueba
    estadistica = varianza_diferencia / varianza_serie
    print(f'\n\nEstadística de la prueba de Dickey-Fuller para {nombre_columna}: {estadistica}')

    # Comparación de la estadística con el valor crítico
    valor_c = 0.01
    if estadistica < valor_c:
        print(f'\nLa serie {nombre_columna} parece ser estacionaria (se rechaza la hipótesis nula)')
    else:
        print(f'No se puede rechazar la hipótesis nula para {nombre_columna}: la serie puede no ser estacionaria\n')

    plt.figure(figsize=(12, 6))
    plt.plot(serie, color='lightsteelblue', label='Serie Temporal')
    plt.plot(media_movil, color='lightpink', label='Media Móvil')
    plt.legend(loc='best')
    plt.title(f'Serie Temporal y Media Móvil - {nombre_columna}')
    plt.show()


# Dickey-Fuller a todas las columnas numéricas
for columna in columnas_numericas:
    serie_temporal = datos[columna]
    prueba_dickey_fuller(serie_temporal, columna)
```
